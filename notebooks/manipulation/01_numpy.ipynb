Dans le monde de la science des donn√©es, comme cela sera √©voqu√©

# 3. Indexation et *slicing*

## 3.1 Logique dans le cas d‚Äôun array unidimensionnel

La structure la plus simple est l‚Äô*array* unidimensionnel:


````python
x = np.arange(10)
print(x)
````

L‚Äôindexation est dans ce cas similaire √† celle d‚Äôune liste:

-   le premier √©l√©ment est 0
-   le √©ni√®me √©l√©ment est accessible √† la position $n-1$

La logique d‚Äôacc√®s aux √©l√©ments est ainsi la suivante :

``` python

```

Avec un *array* unidimensionnel, l‚Äôop√©ration de *slicing* (garder une coupe du *array*) est tr√®s simple.
Par exemple, pour garder les *K* premiers √©l√©ments d‚Äôun *array*, on fera:


En l‚Äôoccurrence, on s√©lectionne le K$^{eme}$ √©l√©ment en utilisant

``` python

```

Pour s√©lectionner uniquement un √©l√©ment, on fera ainsi:


````python
x = np.arange(10)
x[2]
````

Les syntaxes qui permettent de s√©lectionner des indices particuliers d‚Äôune liste fonctionnent √©galement
avec les *arrays*.

# 5. Simulations num√©riques

`Numpy` est incontournable d√®s lors qu‚Äôon effectue des simulations al√©atoires, ce qui est tr√®s commun en statistiques computationnelles avec un ensemble de m√©thodes dites de Monte-Carlo. Le principe g√©n√©ral est de remplacer un calcul th√©orique difficile (une int√©grale, une probabilit√©, une esp√©rance) par une approximation num√©rique obtenue en r√©p√©tant un grand nombre de tirages al√©atoires et en moyennant la quantit√© d‚Äôint√©r√™t, en s‚Äôappuyant sur la loi des grands nombres et le th√©or√®me central limite pour quantifier la pr√©cision de l‚Äôestimation.

Illustrons empiriquement quelques th√©or√®mes incontournables de la statistique par une s√©rie d‚Äôexercices. Cela permettra d‚Äôexplorer :

-   La loi des grands nombres (**?@tip-exo2fr**);
-   Le th√©or√®me central limite et sa version particuli√®re dans le cas de lancer de pi√®ce, le [th√©or√®me de Moivre-Laplace](https://fr.wikipedia.org/wiki/Th%C3%A9or%C3%A8me_de_Moivre-Laplace) ;
-   Les intervalles de confiance th√©oriques et leur contrepartie empirique √† travers le *bootstrap* ;
-   Le principe des m√©thodes de Monte Carlo avec l‚Äôalgorithme du rejet.

Nous allons avoir besoin des √©l√©ments suivants pour initialiser notre processus g√©n√©rateur de donn√©es.


````python
import numpy as np

def generate_grid(size_max=1000):
    n_small = np.arange(3, 200, 2)
    n_log = np.unique(np.round(np.logspace(np.log10(200), np.log10(size_max), 120)).astype(int))
    return np.unique(np.concatenate([n_small, n_log]))

N_max = 100_000
rng = np.random.default_rng(seed=123)
grid = generate_grid(size_max=N_max)
````

Ces √©l√©ments nous permettent de g√©n√©rer une s√©rie al√©atoire d‚Äôobservations jusqu‚Äô√† une taille `N_max`, puis d‚Äôobserver l‚Äô√©volution des estimateurs empiriques de moments (moyenne et variance) lorsque l‚Äôon ne conserve que les `n` premi√®res observations, pour diff√©rents `n` donn√©s par `grid`. L‚Äôobjectif est d‚Äôillustrer la convergence des estimateurs empiriques vers leurs valeurs th√©oriques lorsque `n` augmente.

Cette relation est le fondement de la science des donn√©es, et plus globalement de la statistique, dans sa dimension inf√©rentielle comme descriptive, puisque sans des intuitions math√©matiques formelles nous aurions du mal √† g√©n√©raliser les interpr√©tations issues de donn√©es observ√©es.

La [loi des grands nombres](https://fr.wikipedia.org/wiki/Loi_des_grands_nombres) peut √™tre illustr√©e assez facilement par le biais de simulations num√©riques.
